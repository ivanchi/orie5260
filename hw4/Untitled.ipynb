```python
from pyspark import SparkConf, SparkContext
import numpy as np
```

```python
conf = SparkConf()
```

```python
sc = SparkContext(conf=conf)
```

```python
centroids = []
data = []
with open("datafile1.txt") as f:
    for i, line in enumerate(f):
        if i == 0:
            k = int(line)
        elif i - 1 in range(k):
            centroids.append(line)
        else:
            data.append(line)
#sc.parallelize(centroids).map(lambda line: np.array([float(x) for x in line.split(' ')])).collect()
```

```python
data = sc.parallelize(data).map(lambda line: np.array([float(x) for x in line.split(' ')])).cache()
```

```python
c1 = sc.parallelize(centroids).map(lambda line: np.array([float(x) for x in line.split(' ')]))
```

```python
dat = data.zipWithIndex()
```

```python
cen = c1.zipWithIndex()
cen.map(lambda line: len(line[0])).collect()
```

```python
distance = dat.cartesian(cen)
```

```python
dist = distance.map(lambda line: (line[0][1], (np.linalg.norm(line[0][0] - line[1][0]), line[0][0], line[1][1])))
dis = dist.reduceByKey(lambda v1, v2: min(v1, v2)).map(lambda line: (line[1][2], (line[1][1], 1)))
dis.take(20)
```

```python
new_cen = dis.reduceByKey(lambda (v1, v2), (x1, x2): (v1 + x1, v2 + x2)).map(lambda l: [l[1][0]/l[1][1]])
new_cen.take(20)
```
